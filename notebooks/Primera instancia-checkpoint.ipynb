```python
# Primer dataset

df_card = pd.read_csv(os.getcwd()+'/data/raw/Dataset1.csv')
df_card.head(10)
```

```python
# Segundo dataset

df_house= pd.read_csv(os.getcwd()+'/data/raw/Dataset2.csv')
df_house.head(10)
```

#### Voy a unir a los dos datasets mediante una clave común. Para crear esta clave común voy a unir 'Gender','Age'y Marital_Status en df_house(el segundo dataset) y sus equivalentes en df_card (primer dataset).

```python
# Normalizamos las columnas Gender
df_card['Gender'] = df_card['Gender'].replace({1: 'M', 0: 'F'}) #Transfomramos en 0 y 1

# Normalizamos las columnas de edad (convertir las edades en enteros)
df_card['Age'] = df_card['Age'].astype(int)
df_house['Customer_Age'] = df_house['Customer_Age'].astype(int)

# Unimos los datasets basados en columnas comunes que mencione antes (Gender, Age, Marital Status)
#'Family_status' en df_card y 'Marital_Status' en df_house se consideran equivalentes

# Combinamos los datasets con las columnas normalizadas
df = pd.merge(df_card, df_house, left_on=['Gender', 'Age', 'Family_status'], right_on=['Gender', 'Customer_Age', 'Marital_Status'], how='inner')

# Primeras filas del dataframe nuevo
df.head()
```
